```python
from abc import ABC, abstractmethod
class Runnable(ABC):

    @abstractmethod
    def invoke(input_data):
        pass
```

```python
import random
class NakliLLM(Runnable):
    def __init__(self):
        print('LLM created')

    def invoke(self, prompt):
        response_list = ['Delhi is the capital of India','IPL is a cricket league','AI means artificial intelligence']
        return {'response': random.choice(response_list)}
```

```python
llm = NakliLLM()
```

```python
class NaklPrompttemplate(Runnable):
    def __init__(self, template, input_variables):
         self.template = template
         self.input_variables = input_variables
        
    def invoke(self, input_dict):
        return self.template.invoke(**input_dict)
    
    
```

```python
class NakilStringoutputparser(Runnable):
    def __init__(self):
        pass

    def invoke(self,input_data):
        return input_data['response']
```
